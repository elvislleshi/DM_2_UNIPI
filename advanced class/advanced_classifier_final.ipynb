```python
# 1. Importiamo pandas (se non è ancora importato)
import pandas as pd

# 2. Selezioniamo tutte le colonne di tipo "object" o "boolean"
# (cioè colonne testuali o con valori True/False)
colonne_categoriali = df.select_dtypes(include=['object', 'bool']).columns.tolist()

# 3. Stampiamo tutte le colonne che richiedono encoding
print("Colonne che richiedono encoding:\n")
for colonna in colonne_categoriali:
    print(f"- {colonna}")
```

```python
 #Extract only the first genre from the 'genres' column
df['main_genre'] = df['genres'].apply(lambda x: str(x).split(',')[0].strip())

# Remove rows where the extracted genre is '\N'
df = df[df['main_genre'] != '\\N'].copy()

# Display unique genre count and top 10 genres
print("Unique genres:", df['main_genre'].nunique())
print("Top 10 most frequent genres:\n", df['main_genre'].value_counts().head(10))
```

```python
# Drop the original 'genres' column
df.drop(columns=['genres'], inplace=True)

# Check that it's gone and preview the dataset
print(df.columns)
df.head()
```

```python
features_to_keep = [
    'rating', 'canHaveEpisodes', 'isAdult',
    'titleType_short', 'titleType_tvEpisode', 'titleType_tvMiniSeries',
    'titleType_tvMovie', 'titleType_tvSeries', 'titleType_tvShort',
    'titleType_tvSpecial', 'titleType_video', 'titleType_videoGame'
]
```

```python
# Set frequency threshold
threshold = 1000

# Find genres below the threshold
rare_genres = df['main_genre'].value_counts()[df['main_genre'].value_counts() < threshold].index

# Replace rare genres with 'Other'
df['main_genre'] = df['main_genre'].apply(lambda x: 'Other' if x in rare_genres else x)

# Check the updated distribution
print("New unique genres:", df['main_genre'].nunique())
print("Genre distribution:\n", df['main_genre'].value_counts())
```

```python
y = df['main_genre']
```

```python
from sklearn.preprocessing import LabelEncoder

# 1. Targeti


# 2. Features pa 'genres' dhe 'isRatable'
features_cleaned = [
    'awardWins', 'totalVideos', 'titleType', 'canHaveEpisodes',
    'isAdult', 'numRegions', 'averageRating',
    'writerCredits', 'directorsCredits', 'durationYears'
]
X_clean = df[features_cleaned].copy()

# 3. Encoding për 'titleType'
le_title = LabelEncoder()
X_clean['titleType'] = le_title.fit_transform(X_clean['titleType'])

# 4. Kontrollim final
print("X_clean shape:", X_clean.shape)
print("y (target) shape:", y.shape)
print("\nShembull të dhënash:")
print(X_clean.head())
print("\nTarget (genres) unik:", y.unique())
```

```python
import numpy as np

# Kontrollojmë që nuk ka asnjë kolonë me tip 'object'
object_columns = X_clean.select_dtypes(include='object').columns.tolist()
if object_columns:
    print("⚠️ Kolona me tip 'object':", object_columns)
else:
    print("✅ Nuk ka kolona me tip 'object'.")

# Kontrollojmë që të gjitha kolonat janë 'int'
non_integer_columns = [col for col in X_clean.columns if not np.issubdtype(X_clean[col].dtype, np.integer)]
if non_integer_columns:
    print("⚠️ Kolona që nuk janë 'int':", non_integer_columns)
else:
    print("✅ Të gjitha kolonat janë 'int'.")

# (Opsionale) Kontrollojmë që canHaveEpisodes ka vetëm 0 dhe 1
if 'canHaveEpisodes' in X_clean.columns:
    unique_vals = X_clean['canHaveEpisodes'].unique()
    if sorted(unique_vals) == [0, 1]:
        print("✅ 'canHaveEpisodes' ka vetëm vlera 0 dhe 1.")
    else:
        print("⚠️ 'canHaveEpisodes' ka vlera të tjera:", unique_vals)
```

```python
# Shndërro të gjitha kolonat në int32
X_clean = X_clean.astype('int32')

# Kontroll i ri pas shndërrimit
object_columns = X_clean.select_dtypes(include='object').columns.tolist()
non_integer_columns = [col for col in X_clean.columns if not np.issubdtype(X_clean[col].dtype, np.integer)]

if not object_columns and not non_integer_columns:
    print("✅ Të gjitha kolonat tani janë të tipit 'int' dhe jo 'object'.")
else:
    print("⚠️ Ende ka kolona të papërshtatshme:")
    if object_columns:
        print(" - Object columns:", object_columns)
    if non_integer_columns:
        print(" - Non-int columns:", non_integer_columns)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Ndarja train/test (me stratifikim për klasat)
X_train, X_test, y_train, y_test = train_test_split(X_clean, y, test_size=0.2, stratify=y, random_state=42)

# Inicializojmë modelin SVM (me kernel linear si fillim)
svm_model = SVC(kernel='linear', C=1, decision_function_shape='ovr')  # ovr = one-vs-rest për multi-class

# Trajnojmë
svm_model.fit(X_train, y_train)

# Parashikojmë
y_pred = svm_model.predict(X_test)

# Vlerësim
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

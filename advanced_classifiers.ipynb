```python
# Set frequency threshold
threshold = 1000

# Find genres below the threshold
rare_genres = df['main_genre'].value_counts()[df['main_genre'].value_counts() < threshold].index

# Replace rare genres with 'Other'
df['main_genre'] = df['main_genre'].apply(lambda x: 'Other' if x in rare_genres else x)

# Check the updated distribution
print("New unique genres:", df['main_genre'].nunique())
print("Genre distribution:\n", df['main_genre'].value_counts())
```

```python
import numpy as np

# 1. Handle 'countryOfOrigin' missing values
df['countryOfOrigin'] = df['countryOfOrigin'].fillna('Unknown')

# 2. Handle 'soundMixes'
# Replace empty lists with NaN
df['soundMixes'] = df['soundMixes'].replace("[]", np.nan)

# Define grouping columns (replacing 'genres' with 'main_genre')
group_cols = ['titleType', 'main_genre', 'startYear']

# Compute the mode for each group
mode_per_group = (
    df.groupby(group_cols)['soundMixes']
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    .reset_index()
    .rename(columns={'soundMixes': 'most_common_mix'})
)

# Merge the mode back to the original dataframe
df = df.merge(mode_per_group, on=group_cols, how='left')

# Fill NaNs in 'soundMixes' with group mode
df['soundMixes'] = df['soundMixes'].fillna(df['most_common_mix'])

# Drop helper column
df.drop(columns=['most_common_mix'], inplace=True)

# 3. Clean 'regions' column
# Remove '\N' entries
df['regions'] = df['regions'].astype(str).str.replace(r"'\\N'\s*,?\s*", "", regex=True)

# Replace empty list representations with ['Unknown']
df['regions'] = df['regions'].replace(r"\[\s*\]", "['Unknown']", regex=True)
```

```python
# Replace empty lists with NaN temporarily
df['soundMixes'] = df['soundMixes'].replace("[]", np.nan)

# Define grouping columns
group_cols = ['titleType', 'main_genre', 'startYear']

# Compute mode per group
mode_per_group = (
    df.groupby(group_cols)['soundMixes']
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    .reset_index()
    .rename(columns={'soundMixes': 'most_common_mix'})
)

# Merge mode back into the original dataframe
df = df.merge(mode_per_group, on=group_cols, how='left')

# Fill NaN with mode; then remaining NaNs with 'Unknown'
df['soundMixes'] = df['soundMixes'].fillna(df['most_common_mix'])
df['soundMixes'] = df['soundMixes'].fillna('Unknown')

# Drop helper column
df.drop(columns=['most_common_mix'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Encode the main_genre column
le = LabelEncoder()
df['genre_encoded'] = le.fit_transform(df['main_genre'])

# Define X and y
X = df.drop(columns=['main_genre', 'genre_encoded'])  # use all other features
y = df['genre_encoded']

# Optional: check class names for reference
genre_classes = le.classes_
print("Encoded genre classes:", genre_classes)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Shapes for confirmation
print("Train set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

# Klasat në të njëjtën renditje si genre_encoded
class_names = ['Action', 'Adult', 'Adventure', 'Animation', 'Biography', 'Comedy',
               'Crime', 'Documentary', 'Drama', 'Family', 'Game-Show', 'Horror',
               'Music', 'Other', 'Reality-TV', 'Short', 'Western']

# Binarizo target-in për multi-class ROC
y_bin = label_binarize(y, classes=np.arange(len(class_names)))
n_classes = y_bin.shape[1]

# Train/test split me stratifikim
X_train, X_test, y_train_bin, y_test_bin = train_test_split(X, y_bin, test_size=0.2, random_state=42, stratify=y)

# Modeli One-vs-Rest SVM me probabilitet (kërkon probability=True)
svm = OneVsRestClassifier(SVC(kernel="linear", C=0.1, probability=True))
svm.fit(X_train, y_train_bin)
y_score = svm.decision_function(X_test)

# ROC për çdo klasë
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Vizato të gjitha ROC curves
plt.figure(figsize=(12, 8))

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], lw=2, label=f"{class_names[i]} (AUC = {roc_auc[i]:.2f})")

plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('🎯 Multi-Class ROC Curve per Genre (SVM One-vs-Rest)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.preprocessing import label_binarize
import pandas as pd
import numpy as np
import itertools
import warnings

# Binarizojmë target-in për ROC AUC
y_bin = label_binarize(y, classes=np.unique(y))
n_classes = y_bin.shape[1]

# Split dataset
X_train, X_test, y_train_bin, y_test_bin = train_test_split(
    X, y_bin, test_size=0.2, stratify=y, random_state=42
)

# Parametrat që duam të testojmë
hidden_layers = [(32,), (32, 32), (64,), (64, 64)]
activations = ['logistic', 'tanh', 'relu']
learning_rates = ['constant', 'adaptive']
solvers = ['sgd', 'adam']

# Lista për të ruajtur rezultatet
results = []

# Testo të gjitha kombinimet
for h, act, lr, solver in itertools.product(hidden_layers, activations, learning_rates, solvers):
    print(f"🔧 Testing MLP with hidden={h}, activation={act}, lr={lr}, solver={solver}")
    try:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")  # Fshih Warning nga konvergjenca

            model = MLPClassifier(
                hidden_layer_sizes=h,
                activation=act,
                learning_rate=lr,
                solver=solver,
                max_iter=300,
                random_state=42
            )
            model.fit(X_train, y_train_bin)
            y_pred_bin = model.predict(X_test)

            # Accuracy (bazuar në klasën më të fortë)
            y_test_labels = np.argmax(y_test_bin, axis=1)
            y_pred_labels = np.argmax(y_pred_bin, axis=1)
            acc = accuracy_score(y_test_labels, y_pred_labels)

            # ROC AUC për klasat (mesatare)
            try:
                auc_avg = roc_auc_score(y_test_bin, model.predict_proba(X_test), average='macro')
            except Exception:
                auc_avg = np.nan  # Disa solverë si 'sgd' nuk mbështesin predict_proba gjithmonë

            # Ruaj rezultatet
            results.append({
                'HiddenLayer': h,
                'Activation': act,
                'LearningRate': lr,
                'Solver': solver,
                'Accuracy': acc,
                'Avg_ROC_AUC': auc_avg
            })

    except Exception as e:
        print(f"❌ Failed: {e}")

# Kthe rezultatet në DataFrame
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='Accuracy', ascending=False)

print("\n✅ ✅ ✅ Tabela përmbledhëse (sorted by Accuracy):")
display(results_df)
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import warnings

# 1. Binarize target for ROC AUC
y_bin = label_binarize(y, classes=np.unique(y))
n_classes = y_bin.shape[1]

# 2. Split data
X_train, X_test, y_train_bin, y_test_bin = train_test_split(
    X, y_bin, test_size=0.2, stratify=y, random_state=42
)
y_train = np.argmax(y_train_bin, axis=1)
y_test = np.argmax(y_test_bin, axis=1)

# 3. Models to evaluate under bagging
models = {
    'SVM (RBF)': SVC(kernel='rbf', C=1, probability=True),
    'Logistic Regression': LogisticRegression(C=1, solver='lbfgs', multi_class='multinomial', max_iter=1000),
    'MLP': MLPClassifier(hidden_layer_sizes=(64,), max_iter=300, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

# 4. Results list
results = []

# 5. Loop through each model
for name, base_model in models.items():
    print(f"\n🔁 Evaluating Bagging - {name}")
    try:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")

            bag_model = BaggingClassifier(
                estimator=base_model,  # 🔧 corrected here
                n_estimators=10,
                random_state=42,
                n_jobs=-1
            )
            bag_model.fit(X_train, y_train)
            y_pred = bag_model.predict(X_test)
            y_proba = bag_model.predict_proba(X_test)

            # Metrics
            acc = accuracy_score(y_test, y_pred)
            prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
            rec = recall_score(y_test, y_pred, average='macro', zero_division=0)
            f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)

            # ROC AUC
            try:
                auc = roc_auc_score(y_test_bin, y_proba, average='macro', multi_class='ovr')
            except:
                auc = np.nan

            results.append({
                'Model': name,
                'Accuracy': acc,
                'Precision': prec,
                'Recall': rec,
                'F1-score': f1,
                'Avg ROC AUC': auc
            })

    except Exception as e:
        print(f"❌ Error with model {name}: {e}")

# 6. Create results DataFrame
results_df = pd.DataFrame(results).sort_values(by='Accuracy', ascending=False)

print("\n📊 Tabela krahasuese e metrikave (Bagging):")
display(results_df)
```
